Problem - 3:

The ’Embarked’ column in the Titanic dataset contains categorical data representing the ports of embarka-
tion:

• ’C’ for Cherbourg
• ’Q’ for Queenstown
• ’S’ for Southampton

```python
 #creates dummy columns from Embarked

embarked_dummies = pd.get_dummies(titanic['Embarked'], prefix="Embarked")

#creates new data frame and adds them
titanic = pd.concat([titanic, embarked_dummies], axis=1)

titanic.drop("Embarked", axis=1, inplace=True)

titanic.head()
```

Problem - 4:

Compare the mean survival rates (’Survived’) for the different groups in the ’Sex’ column. Draw a visual-
ization to show how the survival distributions vary by gender.

```python
titanic.groupby("Sex")["Survived"].mean()
```

visualization

```python
titanic.groupby("Sex")["Survived"].mean().plot(kind='bar')

plt.title("Survival Rate of male / female")
plt.ylabel("Probability of survival")
plt.show()
```

Problem - 5:

Draw a visualization that breaks your visualization from Exercise 3 down by the port of embarkation (’Em-
barked’). In this instance, compare the ports ’C’ (Cherbourg), ’Q’ (Queenstown), and ’S’ (Southampton).

```python
titanic.groupby("Embarked_Q")["Survived"].mean()
titanic.groupby("Embarked_S")["Survived"].mean()
titanic.groupby("Embarked_C")["Survived"].mean()
```

visualization


```python
survival = [
    titanic.groupby("Embarked_C")["Survived"].mean()[1],
    titanic.groupby("Embarked_Q")["Survived"].mean()[1],
    titanic.groupby("Embarked_S")["Survived"].mean()[1]
]

ports = ["Cherbourg", "Queenstown", "Southampton"]

plt.bar(ports, survival)
plt.ylabel("Survival Rate")
plt.title("Survival Rate by Embarkation Port")
plt.show()
```
