SECTION 3.2: EXPLORATORY DATA ANALYSIS PRACTICE

Which passenger had the highest fare paid relative to their age?

```python

print("\n--- Step 1: Adding fare_per_year column ---")
titanic['fare_per_year'] = titanic['Fare'] / titanic['Age']
print("✓ fare_per_year column added")
print(titanic[['Name', 'Fare', 'Age', 'fare_per_year']].head())


print("\n--- Step 2: Filtering fare_per_year > 5 ---")
high_fare_age = titanic[titanic['fare_per_year'] > 5]
print(f"Number of passengers: {len(high_fare_age)}")


print("\n--- Step 3: Sorting by fare_per_year (descending) ---")
high_fare_age_srt = high_fare_age.sort_values('fare_per_year', ascending=False)
print("✓ Sorted successfully")


print("\n--- Step 4: Selecting Name and fare_per_year columns ---")
result = high_fare_age_srt[['Name', 'fare_per_year']]


print("\n--- Step 5: RESULT - Top 10 Passengers with Highest Fare per Year ---")
print(result.head(10))

print(" ANSWER:")
top_passenger = result.iloc[0]
print(f"   Passenger: {top_passenger['Name']}")
print(f"   Fare per Year: ${top_passenger['fare_per_year']:.2f}")
```

Which adult male paid highest fare relative to their class?

```python


# Step 1: Add fare_per_class column (Fare / Pclass)
print("\n--- Step 1: Adding fare_per_class column ---")
titanic['fare_per_class'] = titanic['Fare'] / titanic['Pclass']
print("✓ fare_per_class column added")
print(titanic[['Name', 'Fare', 'Pclass', 'fare_per_class']].head())

# Step 2: Subset for male adults (Age >= 18 and Sex = 'male')
print("\n--- Step 2: Filtering for Adult Males (Age >= 18 and Sex = 'male') ---")
adult_males = titanic[(titanic['Sex'] == 'male') & (titanic['Age'] >= 18)]
print(f"Number of adult male passengers: {len(adult_males)}")

# Step 3: Sort by descending fare_per_class
print("\n--- Step 3: Sorting by fare_per_class (descending) ---")
adult_males_srt = adult_males.sort_values('fare_per_class', ascending=False)
print("✓ Sorted successfully")

# Step 4: Select only Name, Age, and fare_per_class columns
print("\n--- Step 4: Selecting Name, Age, and fare_per_class columns ---")
result = adult_males_srt[['Name', 'Age', 'fare_per_class']]

# Step 5: Look at the result
print("\n--- Step 5: RESULT - Top 10 Adult Males with Highest Fare per Class ---")
print(result.head(10))

print("\n✓ ANSWER:")
top_male = result.iloc[0]
print(f"   Passenger: {top_male['Name']}")
print(f"   Age: {top_male['Age']:.0f}")
print(f"   Fare per Class: ${top_male['fare_per_class']:.2f}")
```
